```python
# 디지털 시계에 하루동안(00:00~23:59) 3이 표시되는 시간을 초로 환산하면 총 몇 초(second) 일까요?
# 디지털 시계는 하루동안 다음과 같이 시:분(00:00~23:59)으로 표시됩니다.
# 00:00 (60초간 표시됨)
# 00:01
# 00:02
# ...
# 23:59
```

```python
sec = 0                    # 시간에 3이 포함되어 있을 시간을 초로 환산
total_sec = 0              # 이러한 sec들의 총합
for i in range(0,24):      # 시간 단위
  for j in range(0,60):    # 분 단위
    if '3' in str(i) or '3' in str(j):        # 시간 혹은 분에 3이라는 글자가 들어가면
      sec = 60 * 60 * i + 60 * j              # 해당 시간을 초 단위로 환산
      total_sec += sec                        # 시간에 더하기
      print('{}:{}, {}, {}'.format(i,j,sec,total_sec))
# total_sec                   #초 출력
```
